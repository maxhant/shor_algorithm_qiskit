This Jupyter notebook helps visualizing the different blocks of Shor's algorithm. They are divided in programmable cells which are aranged such that the functions increase in complexity. It is such that the computer does not have the ressources to compute and test the last two blocks.

<h1> Sum function </h1>
<h3> Principle </h3>
One can run the following cell. It shows the principles behind the Carry operator. 
It will add two qubits together. 
In the output, only the sum of this operation is returned as the result. The excess called the carry is not returned. There are three scenarios: 
<center>
$0 + 0 = 0$ <br>
$1 + 0 = 1$ <br>
$1 + 1= 0$ <br>
</center>
Note that a qubit can be in both states (0 and 1), but it would not allow us to verify the function's working. 
<h3> Test </h3>
The two qubits can be modified in the next cell. Once this is done, the run button (in the top bar) should be pressed.
The corresponding circuit is also represented. 

```python
from visualization_gates import visualization_Sum
#Modify here
qubit1 = 0
qubit2 = 1
#Expected result
print(f"The expected output is {(qubit1+qubit2)%2}.\n")
#Result
c = visualization_Sum(qubit1, qubit2)
c.draw(output='mpl', scale=1, filename='circuit.png', interactive=False, plot_barriers=True, reverse_bits=False, justify=None, vertical_compression='high', idle_wires=True, with_layout=True, fold=-1, ax=None, initial_state=True, cregbundle=True)
```

<h1> Carry function </h1>
<h3> Principle </h3>
One can run the following cell. It shows the principles behind the Carry operator. 
It will add three qubits together with one of them being a carry qubit. 
In the output, the carry of this operation is returned with the result. There are four scenarios: 
<center>
$0 + 0 + 0 = 0 \quad\text{carry}\quad 0$ <br>
$1 + 0 + 0 = 1 \quad\text{carry}\quad 0$ <br>
$1 + 1 + 0= 0 \quad\text{carry}\quad 1$ <br>
$1 + 1 + 1= 1 \quad\text{carry}\quad 1$ <br>
</center>
<h3> Test </h3>
The three qubits can be modified in the cell. Once this is done, the run button (in the top bar) must be pressed.
The corresponding circuit is also represented. 

```python
from visualization_gates import visualization_Carry
#Modify here
carry = 0
qubit1 = 1
qubit2 = 1
#Expected result
summed = qubit1+qubit2+carry
if (summed)>1:
    carry_output = 1
else:
    carry_output = 0
print(f"The expected outputs are {summed%2} with a carry {carry_output}.\n")
#Result
c = visualization_Carry(carry, qubit1, qubit2)
c.draw(output='mpl', scale=1, filename='circuit.png', interactive=False, plot_barriers=True, reverse_bits=False, justify=None, vertical_compression='high', idle_wires=True, with_layout=True, fold=-1, ax=None, initial_state=True, cregbundle=True)
```

<h1> Adder modulo function </h1>
<h3> Principle </h3>
One can run the following cell. It shows the principles behind the Adder Modulo operator. 
It will convert the given numbers into qubit sequences, and then will perform the adder modulo. This operation is first a simple addition, and then it returns the remainder of the sum's division. The dividing number is called the modulo. 
In the output, the result of this two consecutive operations is given. 
<img src ="files/adder_mod.PNG" />
<h3> Test </h3>
The two numbers and the modulo can be modified in the cell. There is one extra input which is number_qubit. It correponds to the number of qubits required to code the number. It must be large enough to encode the numbers, but not too much as it would take too much ressources. 

Once this is done, the run button (in the top bar) must be pressed.
The corresponding circuit is also represented. It is getting large, thus to zoom in, double clicking on the image does it. 

```python
from visualization_gates import visualization_Adder_mod
#Modify here
number1 = 6
number2 = 3
modulo = 12
number_qubits = 4
#Expected result
print(f"The expected output is {(number1+number2)%modulo}.\n")
#Result
c = visualization_Adder_mod(number1, number2, modulo, number_qubits)
c.draw(output='mpl', scale=1, filename='circuit.png', interactive=False, plot_barriers=True, reverse_bits=False, justify=None, vertical_compression='high', idle_wires=True, with_layout=True, fold=-1, ax=None, initial_state=True, cregbundle=True)
```
